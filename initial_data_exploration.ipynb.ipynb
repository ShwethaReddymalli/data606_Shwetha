### We will be adressing the below points in our anlaysis:

1. Identify whether the risk of being affected by lung cancer is higher in Men or Women (gender comparison of data)?
2. Does smoking frequency lead to a higher rate of getting affected?
3. Do older adults affect with cancer experience tendencies of weight loss, shortness of breath, and fatigue?

We will perform indexing of categorical values, visualization and correlation. We are going to split the data in 75 and 25, that indicates training and testing data respectively.

As an extension to our above analysis, we will be starting off with indexing.

```python
#Indexing Categorical Values
lol = df_cancer['Level'].astype('category')
df_cancer['Indexed_Level']=lol.cat.codes
df_cancer.drop(labels=['Level'], axis=1, inplace=True)
df_cancer['Indexed_Level'].value_counts()
#0-High
#1-Low
#2-Medium
```

Reference for Indexing Categorical Values: https://www.datacamp.com/community/tutorials/categorical-data
        

```python
df_cancer.head(5)
```

### Visualizing the indexed categorical values

```python
df_cancer.groupby('Indexed_Level')['Indexed_Level'].count().plot(kind='bar',figsize = (6,6),color = 'orange')
plt.xticks(np.arange(3),("High","Low","Medium"))
plt.show()
```

Reference: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html
           https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xticks.html

We tried to plot the indexed values using a bar chart for understanding the level using the specifications of high, low and medium.

```python
df_cancer.groupby(['Gender','Indexed_Level'])['Gender'].count().unstack('Indexed_Level').plot(kind='bar', stacked=True, color=("green", "pink", "orange"), figsize=(7,7))
plt.xticks(np.arange(2),("Male","Female"))
plt.show()
```

Reference: https://matplotlib.org/stable/gallery/lines_bars_and_markers/bar_stacked.html

Through the analysis we can conclude that men have higher prone of frequency than that of women.

```python
A = df_cancer[df_cancer['Indexed_Level']==0]
B = df_cancer[df_cancer['Indexed_Level']==1]
C = df_cancer[df_cancer['Indexed_Level']==2]
```

```python
#Using Bubbleplot
x= A['chronic Lung Disease']
y= A.Smoking


sns.scatterplot(data=B, x="chronic Lung Disease", y="Smoking", size="chronic Lung Disease", legend=False, sizes=(30, 2500), color = "purple")

sns.scatterplot(data=C, x="chronic Lung Disease", y="Smoking", size="chronic Lung Disease", legend=False, sizes=(30, 2500), color="yellow")

sns.scatterplot(data=A, x="chronic Lung Disease", y="Smoking", size="chronic Lung Disease", legend=False, sizes=(30, 2500), color="red")

plt.rcParams["figure.figsize"] = (6,7)

plt.show()
```

Reference : https://towardsdatascience.com/bubble-plots-in-matplotlib-3f0b3927d8f9

From bubble plot we can understand that a higher smoking tendency leads to higher chances of lung cancer!

```python
correlation_matrix = df_cancer.corr(method='spearman')

fig = plt.figure(figsize=(20,20))
sns.heatmap(correlation_matrix,vmin=-1,vmax=+1,cmap='copper_r',annot=True, 
            linewidths=3,linecolor = 'white')
plt.title('correlation_matrix')
plt.show()
```

Reference: https://vitalflux.com/correlation-heatmap-with-seaborn-pandas/

```python
#Using Pairplot
columns=['Air Pollution','Alcohol use','Dust Allergy','Genetic Risk','Smoking','Dry Cough','Indexed_Level']
sns.pairplot(data=df_cancer[columns], hue="Indexed_Level",diag_kind = "kde", kind = "scatter", height = 2.5, palette = "dark:salmon_r")
```
